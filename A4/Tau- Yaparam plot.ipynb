```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''calculate simple values of opticla depth from the monte_carlo sim
Obviously the data will be different each run because of the random method to monte carlo'''

tau = np.array([0.1, 0.5 ,1.0 ,1.5,3.0])
compt_param = np.array([2.152e-03,1.032e-02,2.528e-02,4.212e-02,1.249e-01])
```

```python
#determine a polynomial to the data, i guess a power of 2.0
coefficients = np.polyfit(tau, compt_param, 2)

# Create a polynomial function using the coefficients
poly_function = np.poly1d(coefficients)
#make a bunch of data points between tau 
x_fit = np.linspace(min(tau), max(tau), 100)

# Calculate corresponding y values using the fitted function
#compare them visally to the data points
y_fit = poly_function(x_fit)
plt.scatter(tau, compt_param, label='Data')
plt.plot(x_fit, y_fit, color='r', label="Polynomial fit") 
plt.xlabel("Optical Depth")
plt.legend()
plt.ylabel("Compton Paramter")
plt.title("Tau Dependendence on Compton Parameter")
plt.show()
print(poly_function)
```

from this is is obvious The compton Paramter has a non-linear dependece on optical depth
