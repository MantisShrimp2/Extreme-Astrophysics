```python
'''Karan Kumar 14906619'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import kn
```

```python
#constants
q = 4.8e-10 #electron charge
me = 9.1e-28 # mass of electron
c = 2.997e10 # Speed of light in vaccum in cm/s
v  = 0.3*c #speed of electron
omega = np.logspace(10,14,1000) #angular frequency hz
#set impact paramter to be b_max
b = 1
```

```python
def pow_spectrum(omega):
    '''Caucluate the bremsstralung power spectrum  for a single electron 
    using the modified bessel function order 1'''
    Bessel = kn(1,(omega*b/v))
    A = (8*q**4)/(3*np.pi*c**3 * me**2 * v**4 )
    power = A*(omega**6)*Bessel
    return power
spectrum = pow_spectrum(omega)

plt.plot(omega, spectrum)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$\omega$ (Hz)')
plt.ylabel('$\\frac{dW}{d\omega}$')
plt.title("Bremsstralung spectrum")
plt.show()
```
