```python
'''Karan Kumar 14906619'''
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#define constants
R = 150*(10**21) # cm size of  cygA
d = 690*(10**24) # cm distance to cygA

#hz define the lowe and upper limits of angular frequency 
boundA  = 10 #use 10 an extra order of magnitude to  avoid diving by zero
boundB = 10**9 #observed frequency
freq = np.linspace(boundA, boundB, 500)
jansky_cov = 10**-26 #covert to milla jansky
```

```python
def cygA_Flux(freq):
    '''Determine the frequency integrated flux of cygA 
    integrate this function to return the total flux
    input:
    freq: array of angular frequency (Hz)
    variables:
    A: contants
    F_v :array  Frequency flux (ergs/(cm^2*s*hz))
    
    return 
    F_v'''
    A = (3.24*R**3)/(3*d**2)
    F_V = A*(freq**(-0.6))
    return F_V
    
F_v = cygA_Flux(freq)
```

```python
#plot the flux spectrum
F_v_jansky = F_v*jansky_cov
plt.plot(freq, F_v_jansky)
plt.xscale("log")
plt.yscale('log')
plt.xlabel('Frequency Hz')
plt.title("Frequncy integrated Flux of CygA")
plt.ylabel('$F_v$ $\\frac{mJy}{HZ}$')
plt.show()
```

```python
#integrate the flux to get total flux and show in mjy units

total_flux, abserr = integrate.quad(cygA_Flux,boundA,boundB)
total_flux_jansky = total_flux*jansky_cov
print("the total flux is "+f"{total_flux:.2e}"+"Hz")
print("the total flux is "+f"{total_flux_jansky:.2e}"+" mJY")
```
